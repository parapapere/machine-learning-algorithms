# **Multi-Class Classification Models Comparison**

This notebook conducts a comprehensive evaluation of various machine learning models for multi-class wine classification. The implemented models include K-Nearest Neighbors (k-NN), One-vs-All Support Vector Machines (SVM), Naive Bayes, Decision Trees, and Support Vector Machines.

Dataset:
Utilizing the scikit-learn library, the wine dataset is loaded f and split into training and test sets. The notebook provides an insightful visualization of the confusion matrices for each model.

Models Testing:



*   ***K-Nearest Neighbors (k-NN):*** A flexible k-NN classifier is trained and   evaluated, with accuracy and a heatmap of the confusion matrix presented.
*   ***One-vs-All Support Vector Machines (SVM):*** The One-vs-All approach is employed with a Support Vector Machine, showcasing accuracy and the corresponding confusion matrix.
* ***Naive Bayes:*** A Gaussian Naive Bayes model is trained and evaluated, and the results are visualized through accuracy and a confusion matrix.

* ***Decision Trees:*** A Decision Tree classifier is implemented, and its performance is evaluated with accuracy and a confusion matrix.

* ***Support Vector Machines (SVM):*** A standard Support Vector Machine is trained and assessed, presenting accuracy and the associated confusion matrix.


The notebook offers a comparative analysis of these models, aiding in the selection of an optimal classifier for the wine classification task.

```python
import pandas as pd
import numpy as np

from statistics import mode

from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

from sklearn.metrics import classification_report
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

#**Dataset**

```python
from google.colab import drive

drive.mount('/drive')
```

```python
train_dataset = pd.read_csv("/drive/My Drive/Colab Notebooks/Wine dataset/train.csv")
test_dataset = pd.read_csv("/drive/My Drive/Colab Notebooks/Wine dataset/test.csv")
```

```python
X_train = train_dataset
y_train = X_train.pop('class')
X_test = test_dataset
y_test = X_test.pop('class')
```

#**Models testing**

##K-nn

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
knn_predictions = knn_model.predict(X_test)
knn_accuracy = accuracy_score(y_test, knn_predictions)
knn_confusion_matrix = confusion_matrix(y_test, knn_predictions)
print("Accuracy:", knn_accuracy*100, "%")
plt.figure(figsize=(8, 6))
sns.heatmap(knn_confusion_matrix, annot=True, cmap="Blues")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.title("Confusion Matrix - k-nn")
plt.show()
```

##One-vs-all

```python
ova_model = OneVsRestClassifier(SVC())
ova_model.fit(X_train, y_train)
ova_predictions = ova_model.predict(X_test)
ova_accuracy = accuracy_score(y_test, ova_predictions)
ova_confusion_matrix = confusion_matrix(y_test, ova_predictions)
print("Accuracy:", ova_accuracy*100, "%")
plt.figure(figsize=(8, 6))
sns.heatmap(ova_confusion_matrix, annot=True, cmap="Blues")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.title("Confusion Matrix - One-vs-All")
plt.show()
```

##Naive Bayes

```python
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
nb_predictions = nb_model.predict(X_test)
nb_accuracy = accuracy_score(y_test, nb_predictions)
nb_confusion_matrix = confusion_matrix(y_test, nb_predictions)
print("Accuracy:", nb_accuracy*100, "%")
plt.figure(figsize=(8, 6))
sns.heatmap(nb_confusion_matrix, annot=True, cmap="Blues")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.title("Confusion Matrix - Naive Bayes")
plt.show()
```

##Decision Tree

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_predictions)
dt_confusion_matrix = confusion_matrix(y_test, dt_predictions)
print("Accuracy:", dt_accuracy*100, "%")
plt.figure(figsize=(8, 6))
sns.heatmap(dt_confusion_matrix, annot=True, cmap="Blues")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.title("Confusion Matrix - Drzewa decyzyjne")
plt.show()
```

##Support vector machine

```python
svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_predictions)
svm_confusion_matrix = confusion_matrix(y_test, svm_predictions)
print("Accuracy:", svm_accuracy*100, "%")
plt.figure(figsize=(8, 6))
sns.heatmap(svm_confusion_matrix, annot=True, cmap="Blues")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.title("Confusion Matrix - Maszyny wektorw nonych (SVM)")
plt.show()
```
